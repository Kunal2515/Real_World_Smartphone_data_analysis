### Q1. What are the top 5 expensive brand?

```python
average_prices = data.groupby('brand_name')['price'].mean().reset_index()
average_prices = average_prices.sort_values(by='price', ascending=False)

top_5_expensive_brands = average_prices.head(5)
sns.barplot(top_5_expensive_brands['brand_name'],average_prices['price'])
```

### Q2. What are the top 5 expensive mondel?

```python
average_prices = data.groupby('model')['price'].mean().reset_index()
average_prices = average_prices.sort_values(by='price', ascending=False)

top_5_expensive_models = average_prices.head(5)
sns.barplot(top_5_expensive_models['model'],average_prices['price'])
plt.xticks(rotation=60)
plt.show()
```

### Q3. Which brand has highest rating?

```python
average_rating = data.groupby('brand_name')['avg_rating'].mean().reset_index()
average_rating = average_rating.sort_values(by='avg_rating', ascending=False)

top_5_expensive_brands = average_rating.head(5)
sns.barplot(top_5_expensive_brands['brand_name'],average_rating['avg_rating'])
```
